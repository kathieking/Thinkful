#### Part 1:  Interpretation and significance

Suppose that we would like to know how much families in the US are spending on recreation annually. We've estimated the following model:

𝑒𝑥𝑝𝑒𝑛𝑑𝑖𝑡𝑢𝑟𝑒 = 873 + 0.0012𝑎𝑛𝑛𝑢𝑎𝑙_𝑖𝑛𝑐𝑜𝑚𝑒 + 0.00002𝑎𝑛𝑛𝑢𝑎𝑙_𝑖𝑛𝑐𝑜𝑚𝑒2 − 223.57ℎ𝑎𝑣𝑒_𝑘𝑖𝑑𝑠

expenditure is the annual spending on recreation in US dollars, annual_income is the annual income in US dollars, and have_kids is a dummy variable indicating the families with children. Interpret the estimated coefficients. What additional statistics should be given in order to make sure that your interpretations make sense statistically. Write up your answer.

𝑦=𝛽0+𝛽1𝑥1+𝛽2𝑥2+...+𝛽𝑛𝑥𝑛+𝜖(𝑒𝑞.1)

target = intercept + features (coefficiencts)+ error (residual)


Y (expenditure) = intercept (+873) + 0.0012𝑎𝑛𝑛𝑢𝑎𝑙_𝑖𝑛𝑐𝑜𝑚𝑒 (coef) + 0.00002𝑎𝑛𝑛𝑢𝑎𝑙_𝑖𝑛𝑐𝑜𝑚𝑒2 (coef) − 223.57ℎ𝑎𝑣𝑒_𝑘𝑖𝑑𝑠 (coef) + error (unexplained feature)

Need to know stastical significance of the features (coef) with t-test & p-value. Specifically, are the coefficients different from zero? It appears that they are, but a t-test is required to show that they are *stastically significant*. 
For purpose of interpreting the model, we will assume that the coefficients are stastically significant.

The intercept or bias of the model is 873. The coefficients 'annual_income' and 'annual_income2' have a positive relationship with the target. The 'have_kids' coefficient has a negative relationship with the targer: these families will spend -$223.57 less than families without kids. 

```python
#from answer
import numpy as np
import matplotlib.pyplot as plt

# X represents annual income
X = np.arange(10000, 50000, 1000)
# Y represents annual expenditure on recreation
# for families with no children
Y = 873 + 0.0012*X + 0.000002*(X**2)
# Y_kids represents annual expenditure on recreation
# for families with children
Y_kids = (873 - 223.57) + 0.0012*X + 0.000002*(X**2)


plt1 = plt.plot(X, Y, label="with no children")
plt2 = plt.plot(X, Y_kids, label="with children")
plt.xlabel("income")
plt.ylabel("recreation expenditure")
plt.title("The relation between income and recreation expenditure")
plt.legend()
plt.show()
```

Build a linear regression model where your **target variable is the difference between the apparenttemperature and the temperature.** 

```python
Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set which includes
X = weather_df[['humidity','windspeed']]

import statsmodels.api as sm
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The t-test and associated p-value help to quantify if/how much an individual feature (coefficient) has impacted the variance in the target feature. The null hypothesis would be that each variable (coefficient) does NOT impact the outcome (target). A lower p-value allows us to reject the null hypothesis, meaning the variable in question DID have an impact on the outcome (target). 

**Are the estimated coefficients statistically significant?**
Yes, the p-value for both coefficients is <0.05, which indicates that they are stastically significant. 

**Are the signs of the estimated coefficients in line with your previous expectations?**
I would expect temperature and humidity to increase and decrease in the same direction, as humidity holds heat in the air. However, the target is the difference between apparent temperature and temperature. In the summer, the apparent temperature can seem higer than the actual temperature and vice versa in the winter. 
For wind speed, I would expect that it would lower the apparent temperature in relation to the actual temperature, in both summer and winter, so the negative correlation makes sense. 

**Interpret the estimated coefficients. What are the relations between the target and the explanatory variables?**
As the humidity and windspeed rise by one point, they lower the target (difference in apparent and actual temperature) by 3.03 and 0.13 degrees, respectively. 

#####  Next, include the interaction of humidity and windspeed to the model above and estimate the model using OLS. 
Are the coefficients statistically significant? Did the signs of the estimated coefficients for humidity and windspeed change? Interpret the estimated coefficients.

```python
weather_df['humd_wind_intact'] = weather_df['humidity'] * weather_df['windspeed']
```

```python

Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set which includes
X = weather_df[['humidity','windspeed', 'humd_wind_intact']]

import statsmodels.api as sm
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

**Are the coefficients statistically significant?**

Yes, the previous coefficients remain stastically significant, as does the new coefficient. 

**Did the signs of the estimated coefficients for humidity and windspeed change? Interpret the estimated coefficients.**

The signs for the coefficients did change: 
original humidity = -3.0292, revised humidity = 0.1775; 
original windspeed = -0.1193, revised windspeed = 0.0905.
Originally these coefficients were negatively correlated, but adding the interaction between them changed both the sign and impact on the outcome. Humidity now will increase the target by 0.18 degrees for every one point increase, and wind speed will increase the target by 0.09 degrees for every one point increase in the coefficient.
